```python
#Написать программу, которая принимает на вход файл с данными о температуре воздуха и влажности в разных городах за определенный период (например, год),
#выводит на экран среднюю температуру и влажность для каждого города за этот период.
#Кроме того, программа должна строить диаграмму рассеяния, показывающую зависимость температуры от влажности для каждого города.
```

```python
#Подключение библиотек
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as mb
```

```python
#Создание фрейма
weather = pd.DataFrame({'date': ['2021-01-01', '2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02', '2021-01-02', '2021-01-03', '2021-01-03', '2021-01-03'],
 'city': ['Moscow', 'London','New York','Moscow', 'London', 'New York', 'Moscow', 'London', 'New York'],
 'temp': [-5, 5, 10, -4, 6, 9, -3, 7, 8],
 'hum': [80, 90, 70, 75, 85, 65, 70, 80,60]})

#Сохранение фрейма в формате csv
weather.to_csv('weather.csv') 
print(weather)
```

```python
#Средняя температура каждого города:
mean_temp_Moscow = sum(weather.loc[weather["city"] == 'Moscow', 'temp']) / len(weather.loc[weather["city"] == 'Moscow', 'temp'])
mean_temp_London = sum(weather.loc[weather["city"] == 'London', 'temp']) / len(weather.loc[weather["city"] == 'London', 'temp'])
mean_temp_New_York = sum(weather.loc[weather["city"] == 'New York', 'temp']) / len(weather.loc[weather["city"] == 'New York', 'temp'])

#Средняя влажность каждого города:
mean_hum_Moscow = sum(weather.loc[weather["city"] == 'Moscow', 'hum']) / len(weather.loc[weather["city"] == 'Moscow', 'hum'])
mean_hum_London = sum(weather.loc[weather["city"] == 'London', 'hum']) / len(weather.loc[weather["city"] == 'London', 'hum'])
mean_hum_New_York = sum(weather.loc[weather["city"] == 'New York', 'hum']) / len(weather.loc[weather["city"] == 'New York', 'hum'])
```

```python
#Создание фрейма для более удобного вывода найденных значений
mean_temp_hum = pd.DataFrame({
 'city': ['Moscow', 'London', 'New York'],
 'mean_temp': [ mean_temp_Moscow,mean_temp_London,mean_temp_New_York],
 'unit_temp': ["C", "C", "C"],
 'mean_hum': [mean_hum_Moscow,mean_hum_London, mean_hum_New_York],
 'unit_hum': ["%", "%", "%"]
})

#Вывод фрейма
print(mean_temp_hum)
```

```python
#Создание списков
mean_temp = [ mean_temp_Moscow, mean_temp_London, mean_temp_New_York]
mean_hum = [mean_hum_Moscow, mean_hum_London, mean_hum_New_York]
city = ['Moscow', 'London', 'New York']

#Определение координат зависимостей
height = mean_temp
weight = mean_hum

#Создание точек диаграммы рассеивания
mb.scatter(mean_temp[0], mean_hum[0], c= "violet", s=100, marker="*")
mb.scatter(mean_temp[1], mean_hum[1], c="darkseagreen", s=100, marker="*")
mb.scatter(mean_temp[2], mean_hum[2], c="burlywood", s=100, marker="*")

#Дополнительное оформление
mb.legend(city)
mb.xlabel("Температура(С)")
mb.ylabel("Влажность (%)")
mb.title("Зависимость температуры от влажности")
mb.show()
```
